## Take an image of your choice. Performs basic operations using SciPy library:

```python
# a) Read the image 
from scipy import misc
import imageio
f = misc.face()
imageio.imsave('face.png', f) # uses the Image module (PIL)

import matplotlib.pyplot as plt
plt.imshow(f)
plt.show()
```

```python
# b) Plot the image with title.

plt.title("Real image")
plt.imshow(f)
plt.show()
```

```python
# c) Rotate the image
from scipy import misc, ndimage
rotate_image = ndimage.rotate(f, 35)
import matplotlib.pyplot as plt
plt.imshow(rotate_image)
plt.title("rotated the image by 35 degree")
```

```python
# d) Flip the image.
rotate=np.flip(f)
plt.title("Flip of image")
plt.imshow(rotate)
plt.show()
```

```python
image2 = ndimage.filters.sobel(f)
plt.imshow(image2)
plt.title("Soble fiters")
plt.show()
```

```python
guss_image = ndimage.gaussian_filter(f, sigma= 20)
plt.title("Gaussian filter used") 
plt.imshow(guss_image)
plt.show()
```

```python
# Apply morphological dilation and erosion with the variation of size: tuple of ints on the image.
img_mor_dil = ndimage.morphology.grey_dilation(f,10)
plt.imshow(img_mor_dil)
plt.show()
img_mor_ero = ndimage.morphology.grey_erosion(f,10)
plt.imshow(img_mor_ero)
plt.show()
```
