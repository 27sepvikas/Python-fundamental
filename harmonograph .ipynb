# Harmonograph 

##  x(t)= A*sin(t*f + p)*e^(-dt)

Where "A" represent Amplitude, "t" represent time, "p" represent phase, "f" represent frequency and "d" represent damping.  

Typicaly harmonograph has two pendulums, Therefore the equation are 

x(t) = A0*sin(t*f+p0)*e^(-d0*t) + A1*sin(t*f+p1)*e^(-d1*t)
y(t) = A2*sin(t*f+p2)*e^(-d2*t) + A3*sin(t*f+p3)*e^(-d3*t)





```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
n = 5000000
t = np.logspace(np.log10(10),np.log10(500),n)
A = [1,5,5,1]
d = [ .09, .81, .108, .027]
f = [20,10,10,20]

x = A[0]*np.sin(t*f[0])*np.exp(-d[0]*t)+ A[1]*np.sin(t*f[1])*np.exp(-d[1]*t)
y = A[2]*np.sin(t*f[2])*np.exp(-d[2]*t)+ A[3]*np.sin(t*f[3])*np.exp(-d[3]*t)
#z = A[4]*np.sin(t*f[4])*np.exp(-d[4]*t)+ A[5]*np.sin(t*f[5])*np.exp(-d[5]*t)
# ploting the equation 

plt.plot(x,y,'k',linewidth = .1)
plt.axis('off')
plt.show()

```
